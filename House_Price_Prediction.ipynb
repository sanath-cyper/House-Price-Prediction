```python
# Define the target variable (y) and features (X)
y = house_data['price']
```

```python
# Create and train the Linear Regression model
from sklearn.linear_model import LinearRegression
model =  LinearRegression()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

# Assuming 'house_data' contains your features and target variable
# Select the features (X) and target variable (y)
X = house_data[['area']]  # Replace 'total_sqft' with your desired feature columns
y = house_data['price']

# Now split the data
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
model.fit(x_train, y_train) # Changed X_train to x_train
```

```python
model.score(x_test, y_test)
```
